## 7. Batch Prediction - Multiple Students

```python
import pandas as pd

# Simulate multiple students
students = [
    {"name": "Alice", "skills": "python sklearn pandas"},
    {"name": "Bob", "skills": "java spring sql"},
    {"name": "Charlie", "skills": "python tensorflow keras"},
    {"name": "Diana", "skills": "javascript react node"},
    {"name": "Eve", "skills": "sql powerbi excel"}
]

# Predict for all students
results = []
for student in students:
    result = predictor.get_career_recommendation(student['skills'])
    results.append({
        'Name': student['name'],
        'Skills': student['skills'],
        'Predicted Career': result['predicted_role'],
        'Coverage %': f"{result['skills_coverage']:.1f}%",
        'Missing Skills Count': len(result['missing_skills'])
    })

# Display as DataFrame
df_results = pd.DataFrame(results)
print("\n" + "="*80)
print("BATCH PREDICTION RESULTS")
print("="*80)
print(df_results.to_string(index=False))
print("="*80)
```
